# Canadian Rental Market Pressure Analysis

In this notebook, I prepare our finalized rental market dataset for import into Power BI. I load the scoring segments table, clean up the city names, and map each to a province code so that Power BI can map and filter the locations correctly.

```python
import pandas as pd
import os

# Load the scored segments table
df = pd.read_csv("../outputs/tables/rental_market_segments.csv")
df.head()
```

## Mapping Provinces and Cleaning City Names

Since the raw CMHC data only has full names like "Vancouver CMA", I map each city to its province and separate the name to get a clean city field and geography type (CMA vs CA).

```python
province_map = {
    "St. John's CMA": "NL",
    "Charlottetown CA": "PE",
    "Halifax CMA": "NS",
    "Saint John CMA": "NB",
    "Moncton CMA": "NB",
    "Fredericton CMA": "NB",
    "Montréal CMA": "QC",
    "Québec CMA": "QC",
    "Sherbrooke CMA": "QC",
    "Trois-Rivières CMA": "QC",
    "Saguenay CMA": "QC",
    "Drummondville CMA": "QC",
    "Ottawa-Gatineau CMA (Qué. part)": "QC",
    "Toronto CMA": "ON",
    "Ottawa-Gatineau CMA (Ont. part)": "ON",
    "Hamilton CMA": "ON",
    "Kitchener-Cambridge-Waterloo CMA": "ON",
    "London CMA": "ON",
    "Windsor CMA": "ON",
    "St. Catharines-Niagara CMA": "ON",
    "Oshawa CMA": "ON",
    "Barrie CMA": "ON",
    "Kingston CMA": "ON",
    "Guelph CMA": "ON",
    "Brantford CMA": "ON",
    "Peterborough CMA": "ON",
    "Thunder Bay CMA": "ON",
    "Belleville - Quinte West CMA": "ON",
    "Greater Sudbury/Grand Sudbury CMA": "ON",
    "Winnipeg CMA": "MB",
    "Regina CMA": "SK",
    "Saskatoon CMA": "SK",
    "Calgary CMA": "AB",
    "Edmonton CMA": "AB",
    "Lethbridge CMA": "AB",
    "Red Deer CMA": "AB",
    "Vancouver CMA": "BC",
    "Victoria CMA": "BC",
    "Kelowna CMA": "BC",
    "Abbotsford-Mission CMA": "BC",
    "Chilliwack CMA": "BC",
    "Kamloops CMA": "BC",
    "Nanaimo CMA": "BC"
}

df["province"] = df["centre"].map(province_map)
df["city"] = df["centre"].apply(lambda x: x.replace(" CMA", "").replace(" CA", ""))
df["geography_type"] = df["centre"].apply(lambda x: "CMA" if "CMA" in x else "CA")

df.head()
```

## Export Dashboard Table

I save the flat table as  in our outputs folder, ready to be loaded into Power BI.

```python
export_path = "../outputs/tables/dashboard_rental_market.csv"
df.to_csv(export_path, index=False)
print(f"Successfully exported {len(df)} rows to: {export_path}")
```
